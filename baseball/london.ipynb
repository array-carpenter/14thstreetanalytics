```python
from scipy.stats import ttest_rel, wilcoxon

# Function to perform statistical tests
def perform_stat_tests(pre_values, post_values):
    ttest_result = ttest_rel(pre_values, post_values)
    wilcoxon_result = wilcoxon(pre_values, post_values)
    return ttest_result.pvalue, wilcoxon_result.pvalue

# Perform statistical tests for each metric
stat_results = {}
for metric in metrics:
    pre_values = [team_stats_df.loc[team]['Pre London'][metric] for team in team_stats_df.index]
    post_values = [team_stats_df.loc[team]['Post London'][metric] for team in team_stats_df.index]
    
    ttest_pvalue, wilcoxon_pvalue = perform_stat_tests(pre_values, post_values)
    stat_results[metric] = {'T-Test p-value': ttest_pvalue, 'Wilcoxon p-value': wilcoxon_pvalue}

# Convert stat_results to DataFrame and display it
stat_results_df = pd.DataFrame(stat_results).T
print(stat_results_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
cardinals_2023_pre = pd.read_csv('cardinals_2023_pre_london_statcast_data.csv')
cardinals_2023_post = pd.read_csv('cardinals_2023_post_london_statcast_data.csv')
cubs_2023_pre = pd.read_csv('cubs_2023_pre_london_statcast_data.csv')
cubs_2023_post = pd.read_csv('cubs_2023_post_london_statcast_data.csv')
yankees_2019_pre = pd.read_csv('yankees_2019_pre_london_statcast_data.csv')
yankees_2019_post = pd.read_csv('yankees_2019_post_london_statcast_data.csv')
redsox_2019_pre = pd.read_csv('redsox_2019_pre_london_statcast_data.csv')
redsox_2019_post = pd.read_csv('redsox_2019_post_london_statcast_data.csv')

# Define a function to filter data for a specific team
def filter_team_data(df, team_abbr):
    return df[(df['home_team'] == team_abbr) | (df['away_team'] == team_abbr)]

# Filter the data for specific teams
cardinals_2023_pre = filter_team_data(cardinals_2023_pre, 'STL')
cardinals_2023_post = filter_team_data(cardinals_2023_post, 'STL')
cubs_2023_pre = filter_team_data(cubs_2023_pre, 'CHC')
cubs_2023_post = filter_team_data(cubs_2023_post, 'CHC')
yankees_2019_pre = filter_team_data(yankees_2019_pre, 'NYY')
yankees_2019_post = filter_team_data(yankees_2019_post, 'NYY')
redsox_2019_pre = filter_team_data(redsox_2019_pre, 'BOS')
redsox_2019_post = filter_team_data(redsox_2019_post, 'BOS')

def calculate_batting_stats(df):
    hits = df[df['events'].isin(['single', 'double', 'triple', 'home_run'])].shape[0]
    at_bats = df[df['events'].notna() & (df['events'] != 'walk')].shape[0]
    walks = df[df['events'] == 'walk'].shape[0]
    total_bases = df['events'].apply(lambda x: 1 if x == 'single' else 2 if x == 'double' else 3 if x == 'triple' else 4 if x == 'home_run' else 0).sum()
    plate_appearances = df.shape[0]

    ba = hits / at_bats if at_bats > 0 else 0
    obp = (hits + walks) / plate_appearances if plate_appearances > 0 else 0
    slg = total_bases / at_bats if at_bats > 0 else 0
    ops = obp + slg

    return {'BA': ba, 'OBP': obp, 'SLG': slg, 'OPS': ops}

def calculate_era(df):
    # Calculate earned runs
    df['earned_run'] = df['events'].apply(lambda x: 1 if x in ['single', 'double', 'triple', 'home_run', 'sac_fly', 'groundout', 'field_out', 'force_out', 'grounded_into_double_play', 'walk'] else 0)
    
    # Calculate outs made by each pitcher
    df['outs'] = df['events'].apply(lambda x: 1 if x in ['strikeout', 'groundout', 'field_out', 'force_out', 'double_play', 'flyout', 'lineout', 'sac_fly', 'pop_out'] else 0)
    
    # Calculate innings pitched
    outs = df['outs'].sum()
    innings_pitched = outs / 3
    earned_runs = df['earned_run'].sum()
    
    # Debugging output
    print(f'Outs: {outs}, Innings Pitched: {innings_pitched}, Earned Runs: {earned_runs}')
    
    # Calculate ERA
    if innings_pitched > 0:
        era = (earned_runs / innings_pitched) * 9
    else:
        era = 0
    return era

teams_pre_post = {
    'Cardinals 2023': (cardinals_2023_pre, cardinals_2023_post),
    'Cubs 2023': (cubs_2023_pre, cubs_2023_post),
    'Yankees 2019': (yankees_2019_pre, yankees_2019_post),
    'Red Sox 2019': (redsox_2019_pre, redsox_2019_post)
}

team_stats = {}

for team, (pre_df, post_df) in teams_pre_post.items():
    pre_batting_stats = calculate_batting_stats(pre_df)
    post_batting_stats = calculate_batting_stats(post_df)
    pre_era = calculate_era(pre_df)
    post_era = calculate_era(post_df)

    team_stats[team] = {
        'Pre London': {**pre_batting_stats, 'ERA': pre_era},
        'Post London': {**post_batting_stats, 'ERA': post_era}
    }

# Convert team_stats to DataFrame
team_stats_df = pd.DataFrame(team_stats).T

# Plotting the stats
metrics = ['BA', 'OBP', 'SLG', 'OPS', 'ERA']

# Set up subplots
fig, axes = plt.subplots(len(metrics), 1, figsize=(10, 25))

for i, metric in enumerate(metrics):
    pre_values = [team_stats_df.loc[team]['Pre London'][metric] for team in team_stats_df.index]
    post_values = [team_stats_df.loc[team]['Post London'][metric] for team in team_stats_df.index]
    teams = team_stats_df.index
    
    axes[i].bar(teams, pre_values, width=0.4, label='Pre London', align='center')
    axes[i].bar(teams, post_values, width=0.4, label='Post London', align='edge')
    
    for j, team in enumerate(teams):
        difference = post_values[j] - pre_values[j]
        axes[i].text(j, max(pre_values[j], post_values[j]) + 0.005, f'{difference:.3f}', ha='center')
    
    axes[i].set_title(metric)
    axes[i].legend(loc='upper center')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from pybaseball import statcast

# Get Statcast data for specific dates
statcast_data = statcast(start_dt="2023-03-31", end_dt="2023-06-23")

# Filter for a specific team (e.g., Cardinals)
team_data = statcast_data[(statcast_data['home_team'] == 'STL') | (statcast_data['away_team'] == 'STL')]

# Identify runs scored
# Assuming that 'events' column indicates scoring events
run_events = ['home_run', 'single', 'double', 'triple']
team_data['runs'] = team_data['description'].str.contains('scores').astype(int)
earned_runs = team_data[team_data['events'].isin(run_events)]['runs'].sum()

# Calculate outs
outs_data = team_data[team_data['description'].str.contains('out|strikeout')]
outs = outs_data['description'].str.contains('out|strikeout').sum()

# Calculate innings pitched
innings_pitched = outs / 3

# Calculate ERA
ERA = (earned_runs / innings_pitched) * 9 if innings_pitched > 0 else 0

print(f'Team ERA: {ERA}')
```
